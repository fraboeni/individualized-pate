```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path

from per_point_pate.data.load_datasets import load_mnist
```

```python
(x, y), shape, classes = load_mnist(Path('../../../../datasets'))
```

```python
fig, axes = plt.subplots(nrows=2, ncols=10, figsize=(10, 4))

for i in range(20):
    ax = axes[i // 10, i % 10]
    img = np.squeeze(x[i])
    ax.imshow(img)
    ax.set_xlabel(y[i])
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0, hspace=-0.5)
plt.savefig(fname='mnist.pdf', dpi=300, format='pdf', transparent=False, bbox_inches=None, pad_inches=0.1)
plt.show()
```
